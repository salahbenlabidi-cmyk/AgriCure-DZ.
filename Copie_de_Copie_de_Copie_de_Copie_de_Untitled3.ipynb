```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 1. تحديد المسار وتجهيز الصور
data_path = '/content/dataset'
train_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    data_path,
    target_size=(224, 224),
    batch_size=2,
    class_mode='categorical'
)

# 2. بناء النموذج البرمجي
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(train_generator.num_classes, activation='softmax')
])

# 3. تجميع النموذج وبدء التدريب
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("⏳ بدأ تدريب الذكاء الاصطناعي الآن...")
model.fit(train_generator, epochs=10)

print("\n✅ تم تدريب النموذج بنجاح على صور الطماطم!")
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

def predict_tomato_health(img_path):
    # 1. تحميل الصورة وتجهيزها (تأكد من مطابقة الحجم 224x224)
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0

    # 2. إجراء التنبؤ
    prediction = model.predict(img_array)

    # 3. عرض النتيجة (الأسماء تعتمد على ترتيب المجلدات لديك)
    classes = ['Healthy (سليمة)', 'Infected (مصابة)']
    result = classes[np.argmax(prediction)]

    print(f"🔍 نتيجة الفحص: {result}")

# ارفع صورة جديدة الآن في قائمة الملفات ثم ضع اسمها هنا لتجربتها
# predict_tomato_health('/content/test_image.jpg')
```
